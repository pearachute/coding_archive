```python
with open('전국렌터카업체정보표준데이터.csv', 'r') as f:
    raw = f.readlines()
    
raw_split = []

for line in raw:
    raw_split.append(line.split(','))
    
#raw_split
    
```

```python
for index, name in enumerate(raw_split[0]): #튜플 형식으로 돌려줌.
    print(index, name)
```

```python
company_address = []
for company in raw_split[1:]:
    company_address.append(company[2])
    
print(company_address)
```

```python
city = []
for i in company_address:
    try:
        city.append(i.split()[0].lstrip('"'))
    except:
        pass
    
city
```

```python
city_set = set(city)
city_set
```

```python
for location in city_set:
    n = city.count(location)
    print(location, n)
```
